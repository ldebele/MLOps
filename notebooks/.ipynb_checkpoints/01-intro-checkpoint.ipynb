```python
df["duration"] = (df.tpep_dropoff_datetime - df.tpep_pickup_datetime)
df.duration = df.duration.apply(lambda td: round(td.total_seconds() / 60, 2))
```

```python
# dropping outliers
df_outliers = df[(df.duration >= 1) & (df.duration <= 60)]

dropped_outliers = round((len(df_outliers) / len(df)) * 100, 2)
df = df_outliers
dropped_outliers
```

```python
# re-cast the ints to strings
categorical = ['PULocationID', 'DOLocationID']
df[categorical] = df[categorical].astype(str)
```

```python
def wrangle(filename):
    # read the datasets
    df = pd.read_parquet(filename)
    
    df["duration"] = (df.tpep_dropoff_datetime - df.tpep_pickup_datetime)
    df.duration = df.duration.apply(lambda td: round(td.total_seconds() / 60, 2))

    # dropping outliers
    df = df[(df.duration >= 1) & (df.duration <= 60)]

    # re-cast the ints to strings
    categorical = ['PULocationID', 'DOLocationID']
    df[categorical] = df[categorical].astype(str)

    return df
```

```python
df = wrangle('../data/yellow_tripdata_2023-01.parquet')
df.head()
```

```python
df_val = wrangle('../data/yellow_tripdata_2023-02.parquet')
df_val.head()
```

```python
categorical = ['PULocationID', 'DOLocationID']
dicts = df[categorical].to_dict(orient='records')

# Initialize
dv = DictVectorizer()

# fit and transform the data
df_matrix = dv.fit_transform(dicts)
```

```python
# dimensionality of this matrix
len(dv.get_feature_names_out())
```

```python
categorical = ['PULocationID', 'DOLocationID']
dicts = df_val[categorical].to_dict(orient='records')

# transform the data
df_val_matrix = dv.transform(dicts)
```

```python
target = "duration"
y = df[target].values
y_val = df_val[target].values
```

### Train the Model

```python
# Train the model
lr = LinearRegression()
lr.fit(df_matrix, y)
```

### Predict

```python
y_pred = lr.predict(df_matrix)
mse = mean_squared_error(y, y_pred)
rmse = math.sqrt(mse)
rmse
```

```python
y_val_pred = lr.predict(df_val_matrix)
mse = mean_squared_error(y_val, y_val_pred)
rmse = math.sqrt(mse)
rmse
```
